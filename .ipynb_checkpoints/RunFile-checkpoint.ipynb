```python
import aes, os
key = os.urandom(16)
initial_value = os.urandom(16)
encrypted = aes.AES(key).encrypt_ctr(b'Hello Assignment Ten', initial_value)
print("Encrypted Meessage: ",encrypted )
print("Decrypted Meessage: ",aes.AES(key).decrypt_ctr(encrypted, initial_value))
# b'Attack at dawn'
```
